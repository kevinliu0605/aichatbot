```python
from google.colab import drive

drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/ESE 577 Project/HandWrittenQ&A.txt'

# Open and read
with open(file_path, 'r', encoding='utf-8') as f:
    lines = f.readlines()

print("File content:")
print("".join(lines[:9]))

# Initialize
qa_pairs_train = []

# Process the file content into question-answer pairs
for i in range(0, len(lines), 3):
    question = lines[i].strip()
    answer = lines[i+1].strip()

    formatted_qa = f"<s>[INS]@AI.<question>{question}</question>[/INS]<answer>{answer}</answer></s>"

    qa_pairs_train.append(formatted_qa)

# Print Length
print(f"\nTotal number of Q&A pairs: {len(qa_pairs_train)}")

# Print formatted Q&A pairs
print("\nFormatted Q&A pairs:")
for qa in qa_pairs_train[:5]:
    print(qa)
    print()

# Save the formatted Q&A pairs into a file
output_file_path = '/content/drive/MyDrive/ESE 577 Project/formatted_qa_pairs.txt'
with open(output_file_path, 'w', encoding='utf-8') as f_out:
    for qa in qa_pairs_train:
        f_out.write(qa + '\n')

# Confirm
print(f"\nFormatted Q&A pairs have been saved to: {output_file_path}")
```
